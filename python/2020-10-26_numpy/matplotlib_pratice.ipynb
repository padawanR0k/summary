```python
import numpy as np

# 데이터 생성
x = np.arange(-5, 5, 0.1)
y1 = x**2 -2
y2 = 20*np.cos(x)**2 # NumPy에서 cos()는 np.cos()으로 입력

plt.figure(1) # 1번 그래프 창을 생성함
plt.plot(x, y1) # 지정된 그래프 창에 그래프를 그림

plt.figure(2) # 2번 그래프 창을 생성함
plt.plot(x, y2) # 지정된 그래프 창에 그래프를 그림

plt.figure(1) # 이미 생성된 1번 그래프 창을 지정함
plt.plot(x, y2) # 지정된 그래프 창에 그래프를 그림

plt.figure(2) # 이미 생성된 2번 그래프 창을 지정함
plt.clf() # 2번 그래프 창에 그려진 모든 그래프를 지움
plt.plot(x, y1) # 지정된 그래프 창에 그래프를 그림

plt.show()
```

```python
import numpy as np

# 데이터 생성
x = np.arange(0, 10, 0.1)
y1 = 0.3*(x-5)**2 + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2 # NumPy에서 sin()은 np.sin()으로 입력
y4 = 10*np.exp(-x) + 1 # NumPy에서 exp()는 np.exp()로 입력


# 색상옵션값
first_line_color = 'm'
second_line_color = 'y'
third_line_color = 'c'
fourth_line_color = 'k'

# 선 스타일 옵션값
first_line_style = '-'
second_line_style = '--'
third_line_style = ':'
fourth_line_style = '-.'


# 2 × 2 행렬로 이뤄진 하위 그래프에서 p에 따라 위치를 지정
plt.subplot(2,2,1) # p는 1
plt.plot(x,y1, linestyle=first_line_style, color=first_line_color)

plt.subplot(2,2,2) # p는 2
plt.plot(x,y2, linestyle=second_line_style, color=second_line_color)

plt.subplot(2,2,3) # p는 3
plt.plot(x,y3, linestyle=third_line_style, color=third_line_color )

plt.subplot(2,2,4) # p는 4
plt.plot(x,y4, linestyle=fourth_line_style, color=fourth_line_color )

plt.show()
```

## 산점도
- 두 개의  요소로 이뤄진 데이터 집합의 관계 시각화시 용이함
    - 키, 몸무계 관계
    - 기온, 아이스크림 판매량 관계


```python
colors = ['r', 'b', 'g', 'c', 'm', 'k', 'y']
x_axis = [(str(i) + '세') for i in range(0, 101) if i % 10 == 0]

def get_p_chart(names):
    seoul_data = pd.read_csv('https://gist.githubusercontent.com/padawanR0k/ccc285fb7aaa94d814f5f859a8e9eac7/raw/fad36756bd0701452755a20af1050e780c5bc684/seoul_local_population_2.csv', index_col='자치구')
    
    for i, name in enumerate(names):
        gu_name = name
        gu = seoul_data.loc[gu_name]['0세':'100세 이상+']
        labels = gu.keys()
        years = [item for i, item in enumerate(gu)]
        plt.scatter(labels, years, s=10, c=colors[i%7])
            
    plt.legend(names, loc='left', markerfirst=True, title_fontsize=5, ncol=2)
    plt.xticks(x_axis)

get_p_chart(['중구', '도봉구'])
```

```python
get_p_chart(['강남구', '강북구'])
# 재산세 1위와 꼴지
# 노인인구는 비슷하나, 생산가능인구 16~65세 차이가 많다. 
```

```python
get_p_chart(['강남구', '구로구' ,'강북구'])
# 재산세 1위, 중간, 꼴지
# 재산세 순위와 생산가능 인구수 순위가 비례함
```

## 히스토그램
- 데이터를 정해진 간격으로 나누어 시각화 
- 주로 통계 분야에서 데이터가 어떻게 분포하는지 알아볼 때 많이 사용한다.
- 용어
    - 변량 (variate)
        - 데이터를 숫자로 표시한 것
    - 계급 (class)
        - 변량을 정해진 간격으로 나눈 구간
    - 계급간 간격 (class width)
        - 계급을 나눈 크기
    - 도수 (frequency)
        - 나눠진 계급에 속하는 변량의 수
    - 도수 분포표 (frequency distribution table)
        - 계급에 도수를 표시한 표

### 인구분포 데이터를 가지고 10대,  20대, ... 90대로 나누어보자


```python
import pandas as pd
import matplotlib.pyplot as plt
colors = ['r', 'b', 'g', 'c', 'm', 'k', 'y']
x_axis = [(str(i) + '세') for i in range(0, 101) if i % 10 == 0]

def get_p_histogram():
    seoul_data = pd.read_csv('https://gist.githubusercontent.com/padawanR0k/ccc285fb7aaa94d814f5f859a8e9eac7/raw/fad36756bd0701452755a20af1050e780c5bc684/seoul_local_population_2.csv', index_col='자치구')
    gu = seoul_data['25세']
    labels = gu.keys()
    # print(labels)
    years = [item for i, item in enumerate(gu)]
    plt.hist(years, label = labels)
        

get_p_histogram() # 서울특별시 각 자치구들의 25세인구수는 약 6천명인 경우가 제일 많다.

"""
    왼쪽 끝에 분포하는 중구의 경우 25세 인구수가  1937명이다
    오른쪽 끝에 분포하는 관악구의 경우 25세 인구수가 12817명이다.
"""
print(f'25세인구가 가장 적은 자치구는 중구, 가장 많은 곳은 관악구이다. 그 차이는 {round(12817/1937, 1)}배나 된다')
```

```python
get_p_chart(['관악구', '중구'])
```

```python
get_p_chart(['관악구'])
```

- 관악구의 경우 젊은 세대들이 비율이 다른곳에 비해 높다.
- 중구는 젊은 세대 비율이 중년세대 비율과 비슷하면서 젊은 세대 인구가 적기 까지하다.
- 시간이 지남에 따라 중구에 사람들의 유입이 적어지게 되면 집값이 떨어질 수도 있지않을까?
- 관악구의 경우, 젊은 세대들이 많기때문에 인구가 그나마  잘 유지될거 같다. 

```python
import pandas as pd
import matplotlib.pyplot as plt
colors = ['r', 'b', 'g', 'c', 'm', 'k', 'y']
x_axis = [(str(i) + '세') for i in range(0, 101) if i % 10 == 0]

def get_p_pie():
    seoul_data = pd.read_csv('https://gist.githubusercontent.com/padawanR0k/ccc285fb7aaa94d814f5f859a8e9eac7/raw/fad36756bd0701452755a20af1050e780c5bc684/seoul_local_population_2.csv', index_col='자치구')
    sums = sorted(seoul_data['합계'])
    labels = seoul_data['합계'].keys()

    plt.figure(figsize=(5, 5))
    plt.pie(sums, labels=labels, autopct='%.1f%%', counterclock=True, startangle=90)
    
        

get_p_pie() # 강동구가 제일 인구수가 많으며, 종로구가 제일 낮다.

```
